### 2. FRECUENCIA CARDIACA: variable 'X' de entrada 

```python
paciente1_cardio=pd.read_csv(lista_pacientes_frecuencia_cardiaca[n])
paciente1_cardio.columns=["Tiempo","Frec card"]
paciente1_cardio.head()
```

Debemos transformar la frecuencia recogida por el weareable de forma que refleje esa frecuencia cardiaca en los tiempos que se corresponden a las etiquetas del sueño. Deben ser compatibles y tener la misma 'shape'.

En primer lugar, seleccionamos las muestras de frecuencia cardíaca tomadas en el momento del estudio de PSG, es decir, los que tienen un tiempo mayor o igual a 0.

```python
paciente1_cardio_class = paciente1_cardio[paciente1_cardio["Tiempo"]>=0]
```

```python
paciente1_cardio_class.head()
```

```python
x=paciente1_cardio_class['Tiempo']
y=paciente1_cardio_class['Frec card']

plt.figure(figsize=(10,7))
plt.plot(x,y, marker='.')

plt.title("Representación frecuencia cardiaca del primer paciente")
plt.xlabel("Tiempo(s)")
plt.ylabel("Frecuencia cardiaca")
```

En el estudio polisomnográfico se extrae una etiqueta del sueño cada 30 s pero en el dispositivo wearable es difícil ver el patrón de intervalo de tiempo que sigue para captar la información; a simple vista, no parece que siga uno determinado.

Aproximadamente, cada 5 muestras recogidas en el weareable se corresponde con 1 muestra del estudio polisomnográfico. Por tanto se decide recoger la frecuencia cardiaca correspondiente a las etiquetas del sueño.

```python
paciente1_cardio_class = paciente1_cardio_class.iloc[[i for i in range(0,len(paciente1_cardio_class),5)]]
```

```python
paciente1_cardio_class_fin = paciente1_cardio_class.iloc[0:len(paciente1_sueño)]
```

```python
#paciente1_cardio_class_fin['Tiempo_class']=paciente1_sueño['Tiempo']
```

```python
paciente1_cardio_class_fin.head()
```

```python
x=paciente1_cardio_class_fin['Tiempo']
y=paciente1_cardio_class_fin['Frec card']

plt.figure(figsize=(10,7))
plt.plot(x,y, marker='.')

plt.title("Representación frecuencia cardiaca del primer paciente")
plt.xlabel("Tiempo(s)")
plt.ylabel("Frecuencia cardiaca")
```
